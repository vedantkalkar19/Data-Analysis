#Data analysis main topic

```python
df.columns
```

```python
#starting with gender

ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python

sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
sales_gen
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)

sns.barplot(x= 'Gender', y= 'Amount' , data = sales_gen)
```

most are female buyers and there purchasing power is higher then men 

#AGE 


```python
ax = sns.countplot(data = df, x= 'Age Group')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x= 'Age Group', hue = 'Gender') #hue to add different gender data as well

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
 #total Amount Vs Age Group
 
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
```

```python
sns.barplot(x= 'Age Group', y= 'Amount', data= sales_age)
```

this implies that most buyers are of age group 26-35 and majority of them are female

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by= 'Orders', ascending=False).head(6)
```

```python
#sns.set(rc={'figure.figsize':(15,5)})   # just for fun statement
```

```python
sns.barplot(data= sales_state, x = 'State', y = 'Orders')
```

#OCCUPATION

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
```

```python
sns.barplot(data=sales_state, x= 'Occupation', y= 'Amount')
```

the most buyers are working in IT, Aviation and Healthcare sectors

#PRODUCT CATEGORY

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(25,8)})

sns.barplot(data = sales_state, x = 'Product_Category', y= 'Amount')
```

most products sold are food, clothing and electronic gadgets

```python
sales_state = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10) #apply head when data takes a lot of time

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data= sales_state, x= 'Product_ID', y= 'Orders')
```

```python
#top 10 most sold products 

fig1, ax1 = plt.subplots(figsize= (12,5))

df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='pie')  #you can change the type of plotting in (kind='')
```

people of age group 26-35 , from UP, Maharashtra, Karnataka working in IT, Healthcare and Aviation are more likely to buy producrs from Food , Clothing and Electronics category
